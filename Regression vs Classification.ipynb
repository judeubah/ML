# Regression vs Classification

## Regression
Regression is where our algorithm will produce results that are continuous and to make predictions we will be working out where our inputs will have us placed on a linear continuum - like a slope

Examples:
- Height of a person based on their nutrition
- Salary based on education and country of origin
- Car's MPG based on size and engine cyclinders

Linear regression is the most common regression algorithm, as it is simple and easy to use. There are better complex models that can "fit" the data better but are less easy to use


## Classification
Classification is to group our data into categories.
Sometimes the categories are relatively simple and binary (1/0 and True/False).

Multi-class classification is similar to the idea of binary classification except instead of their being two possible outcomes there are three+
For example it might be to predict if an image contains an apple, pear or banana.

Another major difference between binary and multi classification is that in multi there can be an item placed into more than one group ie if the photo had apples, pears and bananas within.

### Methods
To perform these classifications we use models like Naive Bayes, K-Nearest Neighbours, SVMs
